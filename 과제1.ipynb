# 문제1-6) '지역별' 컬럼을 '시도'와 '구군'으로 구분해서 별도의 컬럼을 만드세요

# 문제1-7) '지역고용통계' 를 위한 전처리 함수를 작성하세요.

````python
def employee_preprocessing(path):
    raw_employee = pd.read_excel(path, skiprows = 1)
    #raw_employee.columns = ['지역명','산업별','고용자수','고용자수(RSE)','빈일자리','빈일자리(RSE)','빈일자리율','빈일자리율(RSE)','입직자','입직자(RSE)','입직률','입직률(RSE)','이직자','이직자(RSE)','이직률','이직률(RSE)']
    employee_data = raw_employee[raw_employee['산업별'] == '전산업']    
    select_col = employee_data[['지역별','전체종사자']]
    select_col.columns = ['지역명','고용자수']
    employee_data = select_col
    
    def province(x):
        temp = x.split(' ')[0]
        if len(temp) == 4:
            return temp[0] + temp[2]
        else:
            return temp[:2]
    
    employee_data['시도'] = employee_data['지역명'].dropna().apply(province)
    employee_data['구군'] = employee_data['지역명'].dropna().apply(lambda x: x.split(' ')[1])
    
    return employee_data
````
